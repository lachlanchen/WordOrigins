```python
# Calculate depth of each node
depths = {}
def calculate_depths(node, graph, depth=0):
    if node in depths:  # Prevent infinite loop in case of a cycle
        return
    depths[node] = depth
    for _, child in graph.out_edges(node):
        calculate_depths(child, graph, depth + 1)

calculate_depths(f"pseudopseudohypoparathyroidism_0", G)

# Calculate the width for each depth
width_for_each_depth = {}
for node, depth in depths.items():
    width_for_each_depth[depth] = width_for_each_depth.get(depth, 0) + 1

# Maximum depth for scaling
max_depth = max(width_for_each_depth.keys())

# Position nodes based on depth and width
pos = {}
angle_step = 360 / sum(width_for_each_depth.values())
current_angle = 0

for depth in range(max_depth + 1):
    nodes_at_depth = [node for node, d in depths.items() if d == depth]
    radius = (depth + 1) * 2  # Adjust spacing based on depth
    for node in nodes_at_depth:
        x = np.cos(np.radians(current_angle)) * radius
        y = np.sin(np.radians(current_angle)) * radius
        pos[node] = (x, y)
        current_angle += angle_step

# Plotting the graph using the calculated positions
plt.figure(figsize=(15, 10))
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color="lightblue", alpha=0.7)
nx.draw_networkx_edges(G, pos, arrowstyle="<|-", arrowsize=30, edge_color="gray")
nx.draw_networkx_labels(G, pos, font_size=20, font_weight="bold")

# Adjust edge label rotation and position nodes' example words
for node, (x, y) in pos.items():
    examples = ", ".join(G.nodes[node].get('example_words', []))
    wrapped_text = textwrap.fill(examples, width=20)
    plt.text(x, y - 0.1, wrapped_text, fontsize=14, ha='center', va='top', wrap=True)

plt.title('Etymology of the Word', fontsize=15)
plt.axis('off')
plt.show()
```
